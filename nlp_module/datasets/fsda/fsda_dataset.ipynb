```python
import pandas as pd

# Carregar os dois datasets
df_usda = pd.read_parquet('train-00000-of-00002.parquet')
df_usda2 = pd.read_parquet('train-00001-of-00002.parquet')
df_usda = pd.concat([df_usda, df_usda2], ignore_index=True)
df_pt = pd.read_csv('fsda_insa_dataset_final.csv')

# Filtrar só as linhas cujas LANGUAGE é 'PT'
df_pt = df_pt[df_pt['LANGUAGE'] == 'PT']

df_pt
```

```python

# 🔵 Passo 2: Manter só as colunas que queremos
cols_to_keep = [
    'FOOD_NAME',
    'FOOD_INGREDIENTS',
    'FOOD_SERVING_SIZE',
    'ENERGY (KCAL)',
    'PROTEIN (G)',
    'TOTAL LIPID (FAT) (G)',
    'CARBOHYDRATE, BY DIFFERENCE (G)'
]

# 🔵 Passo 3: Ajustar datasets em português
df_pt['FOOD_SERVING_SIZE'] = "100g"
df_pt['LANGUAGE'] = 'PT'
df_pt = df_pt[cols_to_keep + ['LANGUAGE']]
```

```python

# 🔵 Passo 4: Ajustar datasets USDA
df_usda['LANGUAGE'] = 'EN'
df_usda = df_usda[cols_to_keep + ['LANGUAGE']]

# 🔵 Passo 5: Concatenar os dois
df_merged = pd.concat([df_usda, df_pt], ignore_index=True)

print(f"Dataset final com {len(df_merged)} linhas")
print(df_merged.sample(5))

# Salvar o datasets final
df_merged.to_parquet('fsda_insa_dataset_final.parquet', index=False)

```
